```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data - titanic train.csv')
df
```

Dealing with the missing values:

```python
df.isna().sum()
```

```python
df.Age.fillna(df.Age.mean(), inplace=True)
```

```python
df.Embarked.fillna('S', inplace=True)
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, columns=['Pclass', 'Sex', 'Embarked'], drop_first=True)
df
```

Now we need to apply feature scaling on our numerical values:

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df['Age'] = sc.fit_transform(df['Age'].values.reshape(891,1))
df['Fare'] = sc.fit_transform(df['Fare'].values.reshape(891,1))
```

```python
df
```

Now we have to train the model:

```python
x_train = ['Age','SibSp','Parch','Fare',
              'Pclass_2', 'Pclass_3', 'Sex_male', 'Embarked_Q','Embarked_S']
y_train = 'Survived'
```

```python
df[x_train]
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(df[x_train], df[y_train])
```

Now its time to predict:

Dealing with missing values:

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df['Age'] = sc.fit_transform(df['Age'].values.reshape(418,1))
df['Fare'] = sc.fit_transform(df['Fare'].values.reshape(418,1))
df
```

```python
x_test = ['Age','SibSp','Parch','Fare','Pclass_2', 'Pclass_3', 'Sex_male', 'Embarked_Q','Embarked_S']
df['Survived']= classifier.predict(df[x_test])
df
```
